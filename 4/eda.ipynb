```python
df_country=pd.read_excel("./Country-Code.xlsx")
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on="Country Code",how="left")
```

```python
final_df.head()
```

```python
country_names=final_df.Country.value_counts().index
country_valu=final_df.Country.value_counts().values
```

```python
plt.pie(country_valu[:3],labels=country_names[:3],autopct="%1.2f%%")
plt.show()
```

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:"Rating Count"})
```

```python
ratings
```

## Concolusion
- 4.5 to 4.9 --> perfect
- 4.0 to 4.4   --> very good
- 3.5 to 3.9 --> good
- 2.5 to 3.4 --> average
- 1.8 to 2.4 --> poor
- 0          --> not rated 

```python
plt.rcParams["figure.figsize"]=(12,6)
sns.barplot(data=ratings,x="Aggregate rating",hue="Rating color",y="Rating Count",palette=["navy","red","orange","yellow","green","blue"])
 
```

```python
sns.countplot(x="Rating color",data=ratings,palette=["navy","red","orange","yellow","green","blue"])
```
